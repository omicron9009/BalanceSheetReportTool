```python
import openpyxl
from openpyxl.styles import Border, Side

# Load the input Excel file
input_wb = openpyxl.load_workbook("C:\\Users\\jadit\\OneDrive\\Desktop\\BalanceSheetProject\\ExcelFiles\\InputForScheduleTemplate\\InputForASchedule.xlsx")
input_ws = input_wb.active
ws = input_ws

# Create a new workbook and select the active worksheet
output_wb = openpyxl.Workbook()
output_ws = output_wb.active

# Take input for name and address
company_name = "Saptaranga Research and Organic Private Limited"
address_line1 = "Plot No 45,"
address_line2 = "Ravindra Nagar P.M.G. Society"
address_line3 = "NOTES ANNEXED TO AND FORMING PART OF ACCOUNTS FOR THE YEAR ENDING 31-Mar-2023"
dateCurr = "As on 31-Mar-2023"
datePrev = "As on 31-Mar-2022"

# Set the title of the sheet
output_ws.title = "Financial Report"

# Set the title and subtitle
output_ws.merge_cells('C1:H1')
output_ws['C1'] = company_name
output_ws.merge_cells('C2:H2')
output_ws['C2'] = address_line1
output_ws.merge_cells('C3:H3')
output_ws['C3'] = address_line2
output_ws.merge_cells('C4:H4')
output_ws['C4'] = address_line3

# Set the headers
output_ws['C6'] = "Note :  Other Current Liabilities"
output_ws['G5'] = "In Rs. hundreds"
output_ws['C7'] = "Current Maturities Of Finance Lease Obligations"
output_ws['C8'] = "Interest Accrued But Not Due On Borrowings"
output_ws['C9'] = "Interest Accrued And Due On Borrowings"                #change here ------------------------
output_ws['C10'] = "Income Received In Advance"
output_ws['C11'] = "Unpaid Dividends"
output_ws['C12'] = "Unpaid Matured Deposits And Interest Accrued Thereon"
output_ws['C13'] = "Unpaid Matured Debentures And Interest Accrued Thereon"
output_ws['C14'] = "Other Payables (Specify Nature)"
output_ws['C15'] = "Application Money Received For Allotment Of Securities \n And Due For Refund And Interest Accrued Thereon"
output_ws['C16'] = "Total"
output_ws['G6'] = dateCurr
output_ws['H6'] = datePrev

# Apply formatting
title_font = openpyxl.styles.Font(size=14, bold=True)
header_font = openpyxl.styles.Font(size=10, bold=True)

for cell in ['C1', 'C2', 'C3', 'C4']:
    output_ws[cell].font = title_font
    output_ws[cell].alignment = openpyxl.styles.Alignment(horizontal='center')

output_ws['G5'].font = header_font
output_ws['G6'].font = header_font
output_ws['H6'].font = header_font
output_ws['C6'].font= header_font

# Adjust column widths
output_ws.column_dimensions['C'].width = 50
output_ws.column_dimensions['G'].width = 15
output_ws.column_dimensions['H'].width = 15

# Define the border style
thin_border = Border(
    left=Side(style='thin'),
    right=Side(style='thin'),
    top=Side(style='thin'),
    bottom=Side(style='thin')
)

# Apply the border to the header cells
for cell in ['C6', 'G5', 'G6', 'H6']:
    output_ws[cell].border = thin_border

# Apply the border to all content cells
max_row = output_ws.max_row
max_col = 1
for row in output_ws.iter_rows(min_row=1, max_row=output_ws.max_row, min_col=1, max_col=output_ws.max_column):
    for cell in row:
        if cell.value is not None:
            if cell.column > max_col:
                max_col = cell.column


for row in range(1, max_row + 1):
    for col in range(3, max_col + 1):
        cell = output_ws.cell(row=row, column=col)
        cell.border = thin_border

# Specify the header name of the column you want to iterate through
header_name = 'Major Head'  # Replace 'Column_Name' with the name of your column header
currval=[]
prevVal=[]
subHeadList=[]
mappedDict={}

SearchMajorHead="Other Current Liabilites" #change here ---------------------------------
count=0
i=0
# Find the index of the column with the specified header name
column_index = None
for col in range(1, ws.max_column + 1):
    if ws.cell(row=1, column=col).value == header_name:
        column_index = col
        break
j=column_index
if column_index is not None:
    # Iterate through each row in the column
    for row in range(2, ws.max_row + 2):  # Start from 2 to skip the header row
        cell_value = ws.cell(row=row, column=column_index).value
        #print(cell_value)  # Do whatever you want with the cell value
        if(cell_value==SearchMajorHead):
            count=count+1
            subHeadList.append( ws.cell(row=row, column=column_index+1).value)

else:
    print(f"Column '{header_name}' not found.")
print(count)
print(subHeadList)

for row in range(2, ws.max_row + 1):  # Start from 2 to skip the header row
    if(ws.cell(row=row, column=j+1).value in subHeadList ):#if (subhead in list)
        mappedDict.update({ws.cell(row=row, column=j+1).value : [ws.cell(row=row, column=j-2).value,ws.cell(row=row, column=j-1).value]})

print(mappedDict)  # Do whatever you want with the cell value
sumOfcurr=0
sumofprev=0
for c in range(1, output_ws.max_column + 1):
    for r in range(1, output_ws.max_row + 1):
        cell_value = output_ws.cell(row=r, column=c).value
        if(cell_value in mappedDict):
            if(mappedDict[cell_value][0]is not None):
                sumOfcurr+=mappedDict[cell_value][0]
            if(mappedDict[cell_value][1]is not None):
                sumofprev+=mappedDict[cell_value][1]
print(sumOfcurr)
print(sumofprev)
# Close the workbook when done
output_wb.save("C:\\Users\\jadit\\OneDrive\\Desktop\\BalanceSheetProject\\GeneratedOutput\\TemplateOutput\\OtherCurrentLiabilities.xlsx") #change here -------------
input_wb.close()
```
