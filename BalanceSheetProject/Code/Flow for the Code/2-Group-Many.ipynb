What is many input problem :- This problem occurs when the trial balance sheet has a high number of inputs , causing the chat gpt api to terminate the resulting output 
before its complete execution , causing errors in formation of dictionary . 
Solution :- divide the number the number of inputs and give them as a seperate prompt to the Gpt model , and then combine the resulting output 

```python
import pandas as pd

# Load the Excel file
input_file_path = "C:\\Users\\jadit\OneDrive\\Desktop\\tribalop\\trialbalance-modified.xlsx"
df = pd.read_excel(input_file_path)

# Number of rows per sheet
rows_per_sheet = 150

# Calculate the number of sheets needed
num_sheets = (len(df) // rows_per_sheet) + 1

# Output file path
output_file_path = "C:\\Users\\jadit\\OneDrive\\Desktop\\tribalop\\TrialBal_divided_modified-2.xlsx"
opf=output_file_path
# Create a Pandas Excel writer using openpyxl as the engine
with pd.ExcelWriter(output_file_path, engine='openpyxl') as writer:
    for i in range(num_sheets):
        start_row = i * rows_per_sheet
        end_row = start_row + rows_per_sheet
        # Select the chunk of rows
        chunk = df.iloc[start_row:end_row]
        # Write the chunk to a new sheet, include headers
        sheet_name = f'Sheet_{i+1}'
        chunk.to_excel(writer, sheet_name=sheet_name, index=False)

print(f'Data has been split into {num_sheets} sheets and stored in {output_file_path}.')
```

```python
import pandas as pd
from openpyxl import load_workbook

# Output file path (the new file created)
output_file_path = opf

# Load the Excel file with multiple sheets into pandas DataFrames
sheets_df = pd.read_excel(output_file_path, sheet_name=None)  # Load all sheets into a dictionary of DataFrames

# Load the Excel file with openpyxl
workbook = load_workbook(output_file_path)
particulars_list = []

# Iterate through each sheet
for sheet_name, df in sheets_df.items():
    sheet = workbook[sheet_name]  # Get the sheet by name
    particulars_col = None
    
    # Find the "Particulars" column
    for col in range(1, sheet.max_column + 1):
        if sheet.cell(row=1, column=col).value == "Particulars":
            particulars_col = col
            break

    # If the "Particulars" column is found, iterate through the rows below it and store the values
    if particulars_col:
        current_particulars = []
        for row in range(2, sheet.max_row + 1):
            cell_value = sheet.cell(row=row, column=particulars_col).value
            drCr=sheet.cell(row=row, column=particulars_col+1).value or None
            if cell_value is not None:  # You can add more checks here if needed
                current_particulars.append(cell_value+"-"+drCr )
        
        # Append the list of particulars for the current sheet to particulars_list
        particulars_list.append(current_particulars)

        # Output the list
        print(current_particulars)
    else:
        print(f"The 'Particulars' column was not found in sheet: {sheet_name}")

    # Additional processing with pandas DataFrame if needed
    print(f"Processing sheet: {sheet_name}")

print("Finished processing all sheets.")
print(len(particulars_list))
```

```python
# Initialize a list to store the multi-line string arrays
multi_line_strings_list = []

# Convert particulars_list into a multi-line string array
multi_line_strings = ['\n'.join(map(str, particulars)) for particulars in particulars_list]

# Store the multi-line string arrays in the list
for multi_line_string in multi_line_strings:
    multi_line_strings_list.append(multi_line_string)
# first_sheet_details = multi_line_strings_list[2]
# print(first_sheet_details)
```

```python
prompt_list=[]
for particular in multi_line_strings_list:
    if(particular is not None):
        BetterPrompt="""
        Classify the following particulars into the given major heads and sub-heads.    
        Particulars: """+particular +""" \n Major Heads: \n"""+AllMajorHeads+"""\nSub Heads: \n"""+AllSubheads+"""
        Output the results as a dictionary where each particular maps to a list with the major head and sub-head.
        consider Particulars as the string Key in double quotes 
        Provide the dictionary output as text only and dont write anything else.
        not even python or json in the box ,
        """
        prompt_list.append(BetterPrompt)
# print(prompt_list[2])
```

```python
import os
import time
from openai import OpenAI
import json
api_key = ''
combined_dict={}
client = OpenAI(
    # This is the default and can be omitted
    api_key=os.environ.get(api_key),
)
for i in prompt_list:
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": i,
            }
        ],
        model="gpt-4o",
        temperature=0,
    )
    res=chat_completion.choices[0].message.content
    print(res)
    # Convert the string to a dictionary
    result_dict = json.loads(res)
    combined_dict.update(result_dict)
    print("waiting and and loading the next process...")
    time.sleep(10)
```

```python
import pandas as pd

# Read the Excel file into a DataFrame
df = pd.read_excel(input_file_path )

# Your dictionary with Particulars as keys and lists [Major Head, Sub Head] as values
your_dictionary = combined_dict

# Function to lookup Major Head and Sub Head values from the dictionary
def lookup_major_sub_head(particular):
    return your_dictionary.get(particular, ['N/A', 'N/A'])

# Add Major Head and Sub Head columns using the lookup function
df['Major Head'], df['Sub Head'] = zip(*df['Particulars'].apply(lookup_major_sub_head))

# Write the updated DataFrame back to the Excel file
df.to_excel("C:\\Users\\jadit\\OneDrive\\Desktop\\BalanceSheetProject\GeneratedOutput\\GroupingOutput\\ManyParticularsOutput2.xlsx", index=False)
```
