```python
import openpyxl
import pandas as pd
# Load the workbook and select the active worksheet
file_path = "C:\\Users\\jadit\\OneDrive\\Desktop\\BalanceSheetProject\\ExcelFiles\\InputForGrouping\\Test-Trialbalance-AddCol\\TestSheet.xlsx"  # Replace with your file path
workbook = openpyxl.load_workbook(file_path)
sheet = workbook.active

# Find the "Particulars" column
particulars_col = None
for col in range(1, sheet.max_column + 1):
    if sheet.cell(row=1, column=col).value == "Particulars":
        particulars_col = col
        break

# If the "Particulars" column is found, iterate through the rows below it and store the values
if particulars_col:
    particulars_list = []
    for row in range(2, sheet.max_row + 1):
        cell_value = sheet.cell(row=row, column=particulars_col).value
        if cell_value is not None:  # You can add more checks here if needed
            particulars_list.append(cell_value)

    # Output the list
    print(particulars_list)
else:
    print("The 'Particulars' column was not found.")
```

```python

# Convert all elements of the list to strings
my_list = [str(item) for item in particulars_list]

# Join the elements of the list using '\n'
multiline_string = '\n'.join(my_list)

print(multiline_string)

particulars=multiline_string

```

```python
import pandas as pd

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path  )

# Your dictionary with Particulars as keys and lists [Major Head, Sub Head] as values
your_dictionary = result_dict

# Function to lookup Major Head and Sub Head values from the dictionary
def lookup_major_sub_head(particular):
    return your_dictionary.get(particular, ['N/A', 'N/A'])

# Add Major Head and Sub Head columns using the lookup function
df['Major Head'], df['Sub Head'] = zip(*df['Particulars'].apply(lookup_major_sub_head))

# Write the updated DataFrame back to the Excel file
df.to_excel("C:\\Users\\jadit\\OneDrive\\Desktop\\BalanceSheetProject\GeneratedOutput\\GroupingOutput\\TestGroupOutput.xlsx", index=False)
```
